```
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
dataset
```

```
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

these items/itemsets appear each at 60% of transactions so they are the most bought as the seller must focus on them

```
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

skirts are always bought with pants
as when buying pants, there is 0.75 chance the buyer will get skirt and/or scarf
when buying scarf , there is 0.75 chance the buyer will get pants

```
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

Here, the lift is bigger than 1 so there is a bigger chance skirt and pants are bought toghether than buying skirt alone

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import math
dataset = pd.read_csv("Market_Basket_Optimisation.csv",header= None)

dataset
```

```
liste = []
for i in range(0, 7501):
    liste.append([str(dataset.values[i,j]) for j in range(0, 20)])
liste
```

```
liste[2].remove("nan")
liste[2].count("nan")
for x in range(len(liste)):
  while liste[x].count("nan")>0:
    liste[x].remove("nan")
liste
```

```
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(liste).transform(liste)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.03)
```

```
frequent_itemsets=apriori(df, min_support=0.03, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.01) # associate itemsets with confidence over 70%.
```

```
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=0.1)
```
